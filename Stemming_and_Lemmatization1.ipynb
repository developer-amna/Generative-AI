```python
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer

# Download required datasets
nltk.download('wordnet')
nltk.download('omw-1.4')
```

# STEMMING :
Stemming is a text preprocessing technique used in Natural Language Processing (NLP). It reduces words to their base or root form by removing suffixes. However, the resulting word may not always be a valid dictionary word.

```python
stemmer = PorterStemmer()

words = ["playing", "running", "happiness", "studies"]

print("Stemming Results:")
for word in words:
    print(word, "→", stemmer.stem(word))
```

**LEMMATIZATION**

Lemmatization is a more advanced technique than stemming. It converts words into their correct base or dictionary form (called lemma) using linguistic rules. It produces meaningful and accurate words.

```python
lemmatizer = WordNetLemmatizer()

words = ["running", "better", "went", "studies"]

print("\nLemmatization Results:")
for word in words:
    print(word, "→", lemmatizer.lemmatize(word, pos='v'))
```

### **EMBEDDING**
Embedding is a technique used to convert words or sentences into numerical vectors so that machines can understand them. In this example, we use CountVectorizer to represent text as numerical data based on word frequency.

```python
sentences = ["I love cats", "I love dogs"]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(sentences)

print("\nFeature Names:", vectorizer.get_feature_names_out())
print("Vector Representation:\n", X.toarray())
```
